### Compile Darknet using Nvidia GPU

````python
%cd darknet
!sed -i 's/OPENCV=0/OPENCV=1/' Makefile
!sed -i 's/GPU=0/GPU=1/' Makefile
!sed -i 's/CUDNN=0/CUDNN=1/' Makefile
!sed -i 's/CUDNN_HALF=0/CUDNN_HALF=1/' Makefile
!make
````

### Configure Darknet network for traning YOLO v3 (Download convolutional layer)

````python
# Copy and rename configure file (.cfg) inside the darnet folder 
!cp cfg/yolov3.cfg cfg/yolov3_training.cfg

# Configure parameters for training model depending on classes
!sed -i 's/batch=1/batch=64/' cfg/yolov3_training.cfg
!sed -i 's/subdivisions=1/subdivisions=16/' cfg/yolov3_training.cfg
!sed -i 's/max_batches = 500200/max_batches = 6000/' cfg/yolov3_training.cfg
!sed -i 's/steps=400000,450000/steps=4800,5200/' cfg/yolov3_training.cfg
!sed -i '610 s@classes=80@classes=3@' cfg/yolov3_training.cfg
!sed -i '696 s@classes=80@classes=3@' cfg/yolov3_training.cfg
!sed -i '783 s@classes=80@classes=3@' cfg/yolov3_training.cfg
!sed -i '603 s@filters=255@filters=24@' cfg/yolov3_training.cfg
!sed -i '689 s@filters=255@filters=24@' cfg/yolov3_training.cfg
!sed -i '776 s@filters=255@filters=24@' cfg/yolov3_training.cfg
````

````python
# Create new file (obj.names) with content one name per line (\n)
!echo -e 'Car\nBus\nTruck' > data/obj.names

# Create new file (obj.data) with content one variable per line (\n)
!echo -e 'classes= 3\ntrain  = data/train.txt\nvalid  = data/test.txt\nnames = data/obj.names\nbackup = /content/drive/MyDrive/yolov3' > data/obj.data
!mkdir data/obj
````

### Extract Images and write image file path into train.txt

````python
# Extract the above formatted images from zip file and move to 'data/obj' directory
!unzip /content/gdrive/MyDrive/yolov3/imagesmlt.zip -d data/obj
````

````python
# We're going to convert the class index on the .txt files. As we're working with only one class, it's supposed to be class 0.
# If the index is different from 0 then we're going to change it.
import glob
import os
import re

txt_file_paths = glob.glob(r"data/obj/*.txt")
for i, file_path in enumerate(txt_file_paths):
    # get image size
    with open(file_path, "r") as f_o:
        lines = f_o.readlines()

        text_converted = []
        for line in lines:
            print(line)
            numbers = re.findall("[0-9.]+", line)
            print(numbers)
            if numbers:

                # Define coordinates
                text = "{} {} {} {} {}".format(0, numbers[1], numbers[2], numbers[3], numbers[4])
                text_converted.append(text)
                print(i, file_path)
                print(text)
        # Write file
        with open(file_path, 'w') as fp:
            for item in text_converted:
                fp.writelines("%s\n" % item)
````

````python
import glob
images_list = glob.glob("data/obj/*.jpg")
print(images_list)

#Create training.txt file
file = open("data/train.txt", "w") 
file.write("\n".join(images_list)) 
file.close() 
````

### Starting the Training using these above  parameters

````python
# Download the pretrained weights of darknet model 53
!wget https://pjreddie.com/media/files/darknet53.conv.74
````

````python
# Start the training using 
# It will take several Hours
!./darknet detector train data/obj.data cfg/yolov3_training.cfg darknet53.conv.74 -dont_show
````
